```python
        
import os
import glob
import pandas as pd
os.chdir("/data2/Common/jbackes/oil")
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
print(all_filenames)
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
```

```python
combined_csv.head()
```

```python
combined_csv.to_csv( "/data2/Common/jbackes/oil/Oil_All.csv", index=False, encoding='utf-8-sig')
```
